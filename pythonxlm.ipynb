```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

 # Extract tables from webpage using Pandas. Retain table number 3 as the required dataframe.

```python
tables = pd.read_html(URL)
df  = tables[3]
df
```

# Replace the column headers with column numbers

```python
df.columns = range(df.shape[1])
```

# 1 Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)

# 2 Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.

# 3 Assign column names as "Country" and "GDP (Million USD)"

```python
#1
df = df[[0,2]]
```

```python
df = df.iloc[1:11:]
```

```python
df.columns = ["Country","GDP(mILLION USD)"]
```

# Load the DataFrame to the CSV file named "Largest_economies.csv"

```python
df.to_csv("Largest_economics.csv")
```
